# data cleaning

```python
df['Worldwide'] = df['Worldwide'].str.replace(',','').astype(int)
df['Domestic'] = df['Domestic'].str.replace(',','').astype(int)
df['Foreign'] = df['Foreign'].str.replace(',','').astype(int)
```

```python
df['Domestic_percent'] = df['Domestic_percent'].str.strip('%').astype(float)
df['Foreign_percent'] = df['Foreign_percent'].str.strip('%').astype(float)
```

```python
df
```

# Distribution of worldwide earnings

```python
plt.figure(figsize=(10,5))
plt.plot(df['Worldwide'])
plt.xlabel('movies worldwide')
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Worldwide'] , kde=True , bins= 30)
```

# Distribution of Domestic vs Foreign Earnings

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Domestic'], color='blue', label='Domestic', kde=True , bins=25)
sns.histplot(df['Foreign'], color='orange', label='Foreign', kde=True , bins=25)
plt.xlabel('Earnings')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

# worldwide rate



```python
descending_worldwide = df[['Release Group','Worldwide']].sort_values(by = 'Worldwide' , ascending=False)
```

```python
descending_worldwide
```

```python
top_worldwide = descending_worldwide.head()
least_worldwide = descending_worldwide.tail()
```

```python
plt.figure(figsize=(10,5))
plt.barh(width = 'Worldwide' , y='Release Group' , data= top_worldwide , color = ['#003f5c','#58508d','#bc5090','#ff6361','#ffa600']  )
plt.title('top world wide')
plt.xlabel('wolrdwide earnings')
plt.ylabel('movies')
```

```python
plt.figure(figsize=(10,5))
plt.barh(width = 'Worldwide' , y='Release Group' , data= least_worldwide , color = ['#003f5c','#58508d','#bc5090','#ff6361','#ffa600']  )
plt.title('least world wide')
plt.xlabel('wolrdwide earnings')
plt.ylabel('movies')
```
